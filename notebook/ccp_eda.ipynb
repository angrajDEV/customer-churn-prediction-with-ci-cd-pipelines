```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/raw.csv')
data.head()
```

```python
data['Churn'].value_counts()
```

```python
data.columns
```

```python
data = data.iloc[:,2:]
data.columns
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
data.dtypes
```

```python
import numpy as np
data['TotalCharges'] = data['TotalCharges'].replace(' ', np.nan)
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')

data['TotalCharges'].astype(float)
```

```python
print("categories in partner ",data['Partner'].unique())
print("categories in Dependents ",data['Dependents'].unique())
print("categories in PhoneService ",data['PhoneService'].unique())
print("categories in MultipleLines ",data['MultipleLines'].unique())
print("categories in InternetService ",data['InternetService'].unique())
print("categories in OnlineSecurity ",data['OnlineSecurity'].unique())
print("categories in OnlineBackup ",data['OnlineBackup'].unique())
print("categories in DeviceProtection ",data['DeviceProtection'].unique())
print("categories in TechSupport ",data['TechSupport'].unique())
print("categories in StreamingTV ",data['StreamingTV'].unique())
print("categories in StreamingMovies ",data['StreamingMovies'].unique())
print("categories in Contract ",data['Contract'].unique())
print("categories in PaperlessBilling ",data['PaperlessBilling'].unique())
print("categories in PaymentMethod ",data['PaymentMethod'].unique())
print("categories in Churn ",data['Churn'].unique())
```

```python
numeric = [numeric for numeric in data.columns if data[numeric].dtype != 'O']
categorical = [categorical for categorical in data.columns if data[categorical].dtype == 'O']

print(f'we have {len(numeric)} numeric columns - ', numeric)
print(f'we have {len(categorical)} categorical columns - ', categorical)
```

```python
sns.boxplot(data=data) # has no outliers 
```

```python
plt.rcParams['figure.figsize'] = (20,5)

plt.subplot(1, 5, 1)
size = data['Churn'].value_counts()
labels = 'No', 'Yes'
color = ['red','green']

plt.pie(size, colors = color, labels = labels,autopct = '.%2f%%')
plt.title('churn', fontsize = 20)
# plt.axis('off')

plt.tight_layout()
plt.grid()
plt.show()
```
